# Data import

```python
ds_test = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\KAGGLE\\STORE SALES\\test.csv")
ds_train = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\KAGGLE\\STORE SALES\\train.csv")
ds_store = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\KAGGLE\\STORE SALES\\stores.csv")
ds_oil = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\KAGGLE\\STORE SALES\\oil.csv")
ds_holidays_events = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\KAGGLE\\STORE SALES\\holidays_events.csv")
ds_transactions = pd.read_csv("C:\\Users\\Matheus Poletto\\Desktop\\Cientista de Dados\\KAGGLE\\STORE SALES\\transactions.csv")
```

```python
ds_train = ds_train.reindex(columns=['sales'] + [col for col in ds_train.columns if col != 'sales'])
ds_train
```

# Yggdrasil Decision Forests

## ML Model

```python
ds_train.info()
```

### Tuner

```python
# Começamos com parâmetros para evitar longo tempo de treinamento inicial
tuner =  ydf.RandomSearchTuner()
tuner.choice("num_trees", [100])
tuner.choice("shrinkage", [0.2])
```

### Model training

```python
ydf_model = ydf.GradientBoostedTreesLearner(
        
    label="sales",
    tuner=tuner,
    task=ydf.Task.REGRESSION,
    
    ).train(ds_train)
```

### Model evaluation

```python
ydf_model.describe()
```

```python
# Gera o analyze
analyze = ydf_model.analyze(ds_train, sampling=0.1)

# Exporta
analyze.to_file("analyze.html")

analyze
```

```python
# Gera o evaluation
evaluate = ydf_model.evaluate(ds_train)

# Gera o conteúdo HTML
html_content = evaluate.html()

# Salva o HTML em um arquivo
with open("evaluate.html", "w", encoding="utf-8") as file:
    file.write(html_content)

evaluate
```

### Predicting

```python
#Predicting
ydf_pred = ydf_model.predict(ds_test)
#ydf_pred = ydf_pred.round(0).astype(int)
ydf_pred[ydf_pred < 0] = 0
ydf_pred
```

### Exporting the model

```python
# Export the result
output = pd.DataFrame({'id': ds_test['id'], 'sales': ydf_pred})
#output.to_csv('store_sales_submission.csv', index=False)
print("Your submission was successfully saved!")
```

### Score was 1.59370
