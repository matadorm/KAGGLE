```python
# Verificar pronomes de tratamento existentes
pronomes = ds_train['Name'].apply(lambda x: x.split(', ')[1].split(' ')[0]).unique().tolist()
pronomes
```

```python
# FunÃ§Ã£o para buscar os pronomes de tratamento e incluir na coluna Title
def substrings_in_string(big_string, substrings):
    for substring in substrings:
        if big_string.find(substring) != -1:
            return substring
    print(big_string)  # Para depuraÃ§Ã£o
    return np.nan  # Retorna NaN caso nÃ£o encontre nenhuma substring
```

```python
# CriaÃ§Ã£o da coluna Title atravÃ©s da funÃ§Ã£o
ds_train['Title'] = ds_train['Name'].map(lambda x: substrings_in_string(x, pronomes))
ds_train
```

```python
ds_test['Title'] = ds_test['Name'].map(lambda x: substrings_in_string(x, pronomes))
ds_test
```

```python
# Incluimos o pronme de tratamento pois acusou que estava faltando acima
pronomes.append('Countess.')
```

You may be interested to know that â€˜Jonkheerâ€™ is a male honorific for Dutch nobility. Also interesting is that I was tempted to just send â€˜Drâ€™ -> â€˜Mrâ€™, but decided to check first, and there was indeed a female doctor aboard! It seems 1912 was further ahead of its time than Doctor Who!

Curious, I looked her up: her name was Dr. Alice Leader, and she and her husband were physicians in New York city.

```python
# FunÃ§Ã£o para substituir os pronomes de tratamento conforme 4 categorias, sendo Mr, Mrs, Miss e Dr

def replace_titles(x):
    title=x['Title']
    if title in ['Don.', 'Major.', 'Capt.', 'Jonkheer.', 'Rev.', 'Col.','Mr.','Sir.', 'Master.']:
        return 'Mr'
    elif title in ['Countess.', 'Mme.','Mrs.', 'Lady.', 'Dona.']:
        return 'Mrs'
    elif title in ['Mlle.', 'Ms.', 'Miss.']:
        return 'Miss'
    elif title =='Dr.':
        if x['Sex']=='Male':
            return 'Mr'
        else:
            return 'Mrs'
    else:
        return title
```

```python
# Substituindo os pronomes de tratamento conforme 4 categorias, sendo Mr, Mrs, Miss e Dr
ds_train['Title']=ds_train.apply(replace_titles, axis=1)
ds_train
```

```python
# VerificaÃ§Ã£o se todos foram substituidos
pronomes_alt = ds_train['Title'].unique()
pronomes_alt
```

```python
ds_test['Title']=ds_test.apply(replace_titles, axis=1)
ds_test
```

```python
# VerificaÃ§Ã£o se todos foram substituidos
pronomes_alt_test = ds_test['Title'].unique()
pronomes_alt_test
```

## Deck

We have a â€˜Cabinâ€™ column not doing much, only 1st class passengers have cabins, the rest are â€˜Unknownâ€™. A cabin number looks like â€˜C123â€™. The letter refers to the deck, and so weâ€™re going to extract these just like the titles.

```python
cabin_list = ds_train['Cabin'].unique().tolist()
cabin_list
```

```python
# Ã‰ composto de 'A', 'B', 'C', 'D', 'E', 'F', 'T', 'G', 'Unknown'

cabin_list_alt = ['A', 'B', 'C', 'D', 'E', 'F', 'T', 'G', 'Unknown']
```

```python
# Alterando no ds_train
ds_train['Deck'] = ds_train['Cabin'].map(lambda x: substrings_in_string(str(x), cabin_list_alt) if pd.notnull(x) else 'Unknown')
ds_train
```

```python
# Alterando no ds_test
ds_test['Deck'] = ds_test['Cabin'].map(lambda x: substrings_in_string(str(x), cabin_list_alt) if pd.notnull(x) else 'Unknown')
ds_test
```

```python
ds_test['Family_Size'] = ds_test['SibSp']+ds_test['Parch']
ds_test
```

```python
# VariÃ¡vel de interaÃ§Ã£o devido ao pclass ser indicativo econÃ´mico
ds_train['Age*Class'] = ds_train['Age']*ds_train['Pclass']
ds_train
```

```python
ds_test['Age*Class'] = ds_test['Age']*ds_test['Pclass']
ds_test
```

## Fare per person

## Tuner
