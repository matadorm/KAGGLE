# Yggdrasil Decision Forests

```python
#tuner = ydf.RandomSearchTuner(num_trials=10, automatic_search_space=True)

tuner =  ydf.RandomSearchTuner()
tuner.choice("shrinkage", [0.2, 0.1, 0.05])
tuner.choice("subsample", [1.0, 0.9, 0.8])
tuner.choice("max_depth", [3, 4, 5, 6])
tuner.choice("num_candidate_attributes_ratio", [0.9])
tuner.choice("validation_ratio", [0.2])
tuner.choice("num_trees", [100, 150,200])
tuner.choice("sparse_oblique_num_projections_exponent", [1.6])
```

```python
ydf_model = ydf.GradientBoostedTreesLearner(
    label='SalePrice', 
    task=ydf.Task.REGRESSION,
    tuner = tuner,
    maximum_training_duration_seconds = 10.0,
    split_axis="SPARSE_OBLIQUE", 
    sparse_oblique_normalization="MIN_MAX",
    categorical_algorithm = "RANDOM",
    use_hessian_gain=True,
    
    ).train(ds_train)
```

```python
ydf_model.describe()
```

```python
ydf_model.analyze(ds_train, sampling=0.1)
```

```python
accuracy_ydf = ydf_model.evaluate(ds_train)

# Show the full evaluation report
print("Full evaluation report:")
accuracy_ydf
```

```python
ydf_pred = ydf_model.predict(ds_test)
ydf_pred
```

```python
# Export the result
output = pd.DataFrame({'Id': ds_test['Id'], 'SalePrice': ydf_pred})
output.to_csv('House_prices_final_submission.csv', index=False)
print("Your submission was successfully saved!")
```
