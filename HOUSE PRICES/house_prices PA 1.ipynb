# Yggdrasil Decision Forests

```python
ydf_model = ydf.GradientBoostedTreesLearner(label='SalePrice', task=ydf.Task.REGRESSION).train(ds_train)
```

```python
ydf_model.describe()
```

```python
ydf_model.analyze(ds_train, sampling=0.1)
```

```python
accuracy_ydf = ydf_model.evaluate(ds_train)

# Show the full evaluation report
print("Full evaluation report:")
accuracy_ydf
```

```python
ydf_pred = ydf_model.predict(ds_test)
ydf_pred
```

```python
# Export the result
output = pd.DataFrame({'Id': ds_test['Id'], 'SalePrice': ydf_pred})
output.to_csv('House_prices_original_submission.csv', index=False)
print("Your submission was successfully saved!")
```
